```python
import pandas as pd
# Keep in mind that the filename will likely be different on your computer (especially later in course)

filename = '/content/drive/MyDrive/Pandas/DataSets/mortgages.csv'

df = pd.read_csv(filename)
df
```

```python
df.head()

```

```python
df = df.rename(columns={'Starting Balance': 'starting_balance',

'Interest Paid': 'interest_paid',

'Principal Paid': 'principal_paid'})

# DataFrame after renaming columns

df.head()
```

```python
df.columns = ['month',

'starting_balance',

'repayment',

'interest_paid',
'principal_paid',

'new_balance',

'mortgage_name',

'interest_rate']

df.head()
```

```python
df = df.drop(columns=['new_balance'])

df.head()
```

```python
del df['starting_balance']

df.head()
```
